```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python

tips=sns.load_dataset("tips")
```

```python
tips
```

# Relplot https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
sns.relplot(x='total_bill',y='tip',data=tips);
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker');
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style="time");
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='size',palette='ch:r=-5,l=.75 ');
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size');
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size',sizes=(15,200));
```

```python
ax=sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
ax1=sns.scatterplot(x='total_bill',y='tip',data=tips,hue='time')
```

```python
ax2=sns.scatterplot(x='total_bill',y='tip',data=tips,style='time')
```

```python
ax2=sns.scatterplot(x='total_bill',y='tip',data=tips,style='time',hue='time')
```

```python
ax3=sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day',style='time')
```

```python
ax4=sns.scatterplot(x='total_bill',y='tip',data=tips,size='size')
```

```python
ax5=sns.scatterplot(x='total_bill',y='tip',data=tips,size='size',hue='size');
```

```python
cmap=sns.cubehelix_palette(dark=0.1,light=0.5,as_cmap=True)
ax6=sns.scatterplot(x='total_bill',y='tip',hue="size",size='size',palette=cmap,data=tips)
```

```python
cmap=sns.cubehelix_palette(dark=0.1,light=0.8,as_cmap=True)
ax8=sns.scatterplot(x='total_bill',y='tip',hue="size",size='size',palette=cmap,data=tips,sizes=(20,200),legend="full")
```

```python
cmap=sns.cubehelix_palette(dark=0.1,light=0.8,as_cmap=True)
ax8=sns.scatterplot(x='total_bill',y='tip',hue="day",size='smoker',palette="Set2",data=tips,legend="full")
```

```python
ax9=sns.scatterplot(x='total_bill',y='tip',s=100,color="0.2",marker="+",data=tips)
```
